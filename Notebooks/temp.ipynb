```python
# load the data from Data/data_v1/25-26/spielplan_2025-26.csv
import pandas as pd
df = pd.read_csv("Data/data_v1/25-26/spielplan_2025-26.csv")
```

```python
# print the first 5 rows
print(df.head())
```

```python
# remove rows where 'home_team' is not Grizzlys Wolfsburg
df = df[df['home_team'] == 'Grizzlys Wolfsburg']
print(df.head())
```

```python
# change the season column values from 2025-2026 to 25-26
df['season'] = df['season'].str.replace('2025-2026', '25-26')
print(df.head())
```

```python
# save the dataframe to Data/data_v1/25-26/spielplan_2025-26_cleaned.csv
df.to_csv("Data/data_v1/25-26/spielplan_2025-26_cleaned.csv", index=False)
```

```python
# see unique values in away_team column
print(df['away_team'].unique())
```

```python
# load the df from Data/data_v1/grizzlys_combined_schedule_sales_v3.csv
df_OG = pd.read_csv("Data/data_v1/grizzlys_combined_schedule_sales_v3.csv")

# see unique values in away_team column
print(df_OG['away_team'].unique())
```

```python
# see if you can find any discrepancies between the two away_team unique values
discrepancies = set(df['away_team'].unique()) ^ set(df_OG['away_team'].unique())
print(discrepancies)
```

```python
# see if there are any new teams in the 25-26 season that were not in the previous seasons
new_teams = set(df['away_team'].unique()) - set(df_OG['away_team'].unique())
print(new_teams)
```

```python
# remove the rows with the new teams
df = df[~df['away_team'].isin(new_teams)]
print(df['away_team'].unique())
```

```python
# save the dataframe to Data/data_v1/25-26/spielplan_2025-26_cleaned.csv
df.to_csv("Data/data_v1/25-26/spielplan_2025-26_cleaned.csv", index=False)
```
